```python
import os
import pandas as pd
```

```python
all_files = os.listdir('../data/raw')
data_dict_ref = {'agecl': 'Age', 'all': 'All Families', 'edcl': 'Education', 
                 'famstruct': 'Family Structure', 'housecl': 'Housing Status',
                 'inccat': 'Income Percentile', 'nwcat': 'Net Worth Percentile',
                 'occat1': 'Employment Status', 'occat2': 'Occupation Type',
                 'racecl4': 'Race'
                }
all_csv_files = []
for f in all_files:
    if f.endswith('.csv'):
        all_csv_files.append(f)
all_csv_files
```

```python
master_df = pd.DataFrame()
mean_df = pd.DataFrame()
median_df = pd.DataFrame()
debt_data = all_csv_files.pop()

for filename in all_csv_files:
    if filename == "z1-nonfin-debt.csv":
        # Skip processing for z1-nonfin-debt.csv
        continue
    category = filename.split('_')[3]
    df = pd.read_csv('../data/raw/'+filename, index_col='year')
    df['Categorization'] = data_dict_ref[category]
    if '_mean' in filename:
        df = df.add_suffix('_mean')
        mean_df = pd.concat([mean_df, df], axis=0)
    elif '_median' in filename:
        df = df.add_suffix('_median')
        median_df = pd.concat([median_df, df], axis=0)
```

```python
master_df.shape, mean_df.shape, median_df.shape
```

```python
master_df = pd.concat([master_df, mean_df, median_df], axis=0)
master_df.drop(columns=['Category_median'], inplace=True)
master_df.rename(columns={"Category_mean": "Category"}, inplace=True)
master_df.drop(columns=['Categorization_median'], inplace=True)
master_df.rename(columns={"Categorization_mean": "Categorization"}, inplace=True)
```

```python
first_column = master_df.pop('Categorization')
master_df.insert(0, 'Categorization', first_column)
```

```python
gdp_df = pd.read_csv('../data/raw/world_gdp/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_4150784.csv',  skiprows=3)
us_gdp_df = gdp_df[gdp_df['Country Name']=='United States']
world_gdp_df = gdp_df[gdp_df['Country Name']=='World']
world_gdp_df
```

```python
master_df.head()
```

```python
master_df['US_GDP'] = None
master_df['World_GDP'] = None
master_df
```

```python
for i in master_df.index:
    master_df.at[i, 'US_GDP'] = us_gdp_df[str(i)].values[0]
    master_df.at[i, 'World_GDP'] = world_gdp_df[str(i)].values[0]
master_df.head()
```

```python
master_df.to_csv('../data/sanitized/Masterdata.csv')
```
